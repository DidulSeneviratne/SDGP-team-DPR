```python
import json
import matplotlib.pyplot as plt
import numpy as np
series = []
file = open('../csv/graph.txt', 'r') #read the previous posture status to display the graph
while file.readline() != "":
    try:
        series.append(int(file.readline()))
    except:
        pass
file.close()
max_v = max(series)
with open('graph.json', 'w') as json_file:
    if max_v == -2: # printing what is the average position status
        rate = print("Your rate is - Poor")
        json.dump(rate, json_file)
    elif max_v == -1:
        rate = print("Your rate is - Bad")
        json.dump(rate, json_file)
    elif max_v == 0:
        rate = print("Your rate is - Average")
        json.dump(rate, json_file)
    elif max_v == 1:
        rate = print("Your rate is - Good")
        json.dump(rate, json_file)
    elif max_v == 2:
        rate = print("Your rate is - Excellent")
        json.dump(rate, json_file)
    length = len(series) # calculate read status array list length for adjusting x axis
    def graphVisualization(x, y):
        x_axis = np.array(range(x))
        y_axis = np.array(y)

        plt.plot(x_axis, y_axis)
        plt.title('Posture Analysis')
        plt.xlabel('seconds')
        plt.ylabel('status\n-2 - Poor, -1 - Bad, 0 - Average, 1 - Good, 2 - Excellent')
        return plt.show()
        with open('graph.json', 'a') as json_file:
            json.dump(x_axis, json_file)
        with open('graph.json', 'a') as json_file:
            json.dump(y_axis, json_file)
graphVisualization(length, series)
```
