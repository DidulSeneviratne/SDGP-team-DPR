```python
import cv2
import mediapipe as mp
import numpy as np
from numpy import loadtxt
import os
import time
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
pose = mp_pose.Pose()
points = mp_pose.PoseLandmark # Landmarks
path1 = "../../dataset/good/2" # enter dataset path
path2 = "../../dataset/bad/2"
path3 = "../../dataset/good/4"
path4 = "../../dataset/bad/4"
trained1 = []
trained2 = []
trained3 = []
trained4 = []
```

```python
trained1 = loadtxt('../csv/good-posture1l.txt')
min_angle1 = int(min(trained1))
max_angle1 = int(max(trained1))
trained2 = loadtxt('../csv/good-posture1r.txt')
min_angle2 = int(min(trained2))
max_angle2 = int(max(trained2))
trained3 = loadtxt('../csv/bad-posture3l.txt')
min_angle3 = int(min(trained3))
max_angle3 = int(max(trained3))
trained4 = loadtxt('../csv/bad-posture3r.txt')
min_angle4 = int(min(trained4))
max_angle4 = int(max(trained4))
```

```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
count = 0

status = [None]*144

for img in os.listdir(path1):
    
            img = cv2.imread(path1 + "/" + img)

            imageWidth, imageHeight = img.shape[:2]

            imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            blackie = np.zeros(img.shape) # Blank image

            results = pose.process(imgRGB)
        
            # Extract landmarks
            try:
                landmarks = results.pose_landmarks.landmark

                # Get coordinates
                left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
                right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
                left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
                right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]

                # Calculate angle
                angle1 = calculate_angle(right_shoulder,right_hip,left_hip)

                # Visualize angle
                cv2.putText(img, str(angle1), 
                               tuple(np.multiply(right_hip, [640, 480]).astype(int)), 
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                    )

                if ((int(angle1) in range(min_angle1, max_angle1))):

                    status[count] = "Good"
                    
                count +=1
            
            except:
                pass

            cv2.imshow("Image", img)

            cv2.imshow("blackie",blackie)

            cv2.waitKey(100)
            
print(status)
            
good = 0
            
for posture in status:
    
    if posture == "Good":
    
        good += 1

length = len(status)

precentage = (good/length)*100

print(precentage)

count = 0

status = [None]*146
            
for img in os.listdir(path2):
    
            img = cv2.imread(path2 + "/" + img)

            imageWidth, imageHeight = img.shape[:2]

            imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            blackie = np.zeros(img.shape) # Blank image

            results = pose.process(imgRGB)
        
            # Extract landmarks
            try:
                landmarks = results.pose_landmarks.landmark

                # Get coordinates
                left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
                right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
                left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
                right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]

                # Calculate angle
                angle1 = calculate_angle(right_shoulder,right_hip,left_hip)

                # Visualize angle
                cv2.putText(img, str(angle1), 
                               tuple(np.multiply(right_hip, [640, 480]).astype(int)), 
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                    )

                if ((int(angle1) in range(min_angle3, max_angle3))):

                    status[count] = "Bad"
                    
                count +=1
            
            except:
                pass

            cv2.imshow("Image", img)

            cv2.imshow("blackie",blackie)

            cv2.waitKey(100)
            
print(status)
            
bad = 0
            
for posture in status:
    
    if posture == "Bad":
    
        bad += 1

length = len(status)

precentage = (bad/length)*100

print(precentage)
```
